```python
import pandas as pd
import numpy as np
#this gives us a list of data frames
movies = pd.read_html('https://en.wikipedia.org/wiki/2016_in_film')
movies[2].head()
```

Vamos pegar um data frame da internet, arrumar e salvar em csv
------

```python
dfm = movies[2]
#primeiro colocamos os nomes nas colunas, mas isso gera o problema de
#tornar a linha 0 redundante
dfm.columns = ['Rank', 'Title', 'Distribution', 'Worldwide gross']
dfm.head()
```

```python
#Usamos o metodo drop para apagar a observacao 0
dfm.drop([0], inplace=True)
dfm.head()
```

```python
#Agora vamos usar a coluna Rank para indexar as observacoes.
dfm.index = dfm.Rank
dfm.head()
```

```python
#E assim apagamos a coluna rank
dfm.drop('Rank', 1, inplace=True)
dfm
```

Temos uns erros na tabela. Certos valores estao na coluna errada
olhando a tabela original vemos que a repeticao da companhia
gerou esse problema.

```python
#Vamos por os valores na coluna certa iterando.
for i, corp in enumerate(dfm['Distribution']):
    if '$' in corp:
        dfm['Worldwide gross'][i] = dfm['Distribution'][i]
        dfm['Distribution'][i] = dfm['Distribution'][i-1]
dfm
```

```python
#Vamos agora transformar os valores da coluna gross em int.

import string
symbols =',$'
dic_trans = {key: None for key in symbols}
translator = str.maketrans(dic_trans)

for i,value in enumerate(dfm['Worldwide gross']):
    gross_in_number = int(value.translate(translator))
    dfm['Worldwide gross'][i] = gross_in_number
dfm
```

```python
#alterando o nome da coluna para facil acesso
dfm.rename(columns ={'Worldwide gross':'Gross'}, inplace=True)
dfm
```

```python
#dados basicos
gross = dfm.gross
print('Looking at the highest-grossing films of 2016 \n the mean gross is ${0} and the std is {1}'.\
      format(np.mean(gross),np.std(gross)))
```

```python
dfm.to_csv('2016movies.csv')
```
